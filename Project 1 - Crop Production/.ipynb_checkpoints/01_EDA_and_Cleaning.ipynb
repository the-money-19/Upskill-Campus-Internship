```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_name = 'datafile (1).csv'
df = pd.read_csv(file_name)

# Display the first 5 rows to make sure it worked
print("--- Data Loaded Successfully ---")
print(df.head())
```

```python
# Rename columns for simpler coding
new_columns = {
    'Cost of Cultivation (`/Hectare) A2+FL': 'Cost_Cultivation_A2_FL',
    'Cost of Cultivation (`/Hectare) C2': 'Cost_Cultivation_C2',
    'Cost of Production (`/Quintal) C2': 'Cost_Production_C2',
    'Yield (Quintal/ Hectare) ': 'Yield_Quintal_Hectare' # This is our Target
}
df.rename(columns=new_columns, inplace=True)

# Check the new, clean column names
print("\n--- New Column Names ---")
print(df.columns)
```

```python
# Check data types and look for missing values (Non-Null Count)
print("\n--- Data Quality Check (df.info()) ---")
df.info()

```

```python
# Group by crop and find the average yield
crop_yield_summary = df.groupby('Crop')['Yield_Quintal_Hectare'].mean().sort_values(ascending=False)
crop_yield_df = crop_yield_summary.reset_index()

# Create a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Crop', y='Yield_Quintal_Hectare', data=crop_yield_df)
plt.title('Average Yield by Crop')
plt.xlabel('Crop')
plt.ylabel('Average Yield (Quintal/Hectare)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Print the top 5 for your report
print("\n--- Top 5 Crops by Average Yield ---")
print(crop_yield_summary.head(5)) 
```

```python
# Group by state and find the average yield
state_yield_summary = df.groupby('State')['Yield_Quintal_Hectare'].mean().sort_values(ascending=False)
state_yield_df = state_yield_summary.reset_index()

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='Yield_Quintal_Hectare', data=state_yield_df)
plt.title('Average Yield by State')
plt.xlabel('State')
plt.ylabel('Average Yield (Quintal/Hectare)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Print the top 5 for your report
print("\n--- Top 5 States by Average Yield ---")
print(state_yield_summary.head(5)) 
```
